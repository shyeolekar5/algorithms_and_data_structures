  # Question Two: Path to Leaves

  Given the `root` of a binary tree, return all root to leaf paths in any order.

  ## Examples

  ### Example 1

  Input: `root = [1, 2, 2, 3, 5, 6, 7]` *What traversal method is this?*

  Output: [[1, 2, 3], [1, 2, 5], [1, 2, 6], [1, 2, 7]]

  ### Example 2

  ![](./images/q1_ex3.png)

  Input: `root = [10, 9, 7, 8]`

  Output: [[10, 7], [10, 9, 8]]

</details>

#### Starter Code for Question 2

```python
# Definition for a binary tree node.
from typing import List
class TreeNode(object):
  def __init__(self, val = 0, left = None, right = None):
    self.val = val
    self.left = left
    self.right = right

#Main function to find all paths 
def bt_path(root: TreeNode) -> List[List[int]]:
    paths = []

    def bt_path_helper(node, current_path):
      if node is None: #base case 
        return 
      
      current_path.append(node.val) #current path to have current node

      #if a node has left/right paths, do the same as above. Traverse recursively 
      if node.left: 
        bt_path_helper(node.left, current_path)
      if node.right:
        bt_path_helper(node.right, current_path)


      #if a node does not have left or right path, it is a leaf. So end and add result to paths. 
      if not node.left and not node.right: 
        paths.append(list(current_path))

      #remove the last value if both left and right have been traversed
      current_path.pop()
    
    bt_path_helper(root, [])
    return paths
    
```

```python
#test example 1
root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(2)
root.left.left = TreeNode(3)
root.left.right = TreeNode(5)
root.right.left = TreeNode(6)
root.right.right = TreeNode(7)
result = bt_path(root)
print(result)
```
