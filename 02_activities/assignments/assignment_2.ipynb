
-   Copy the solution your partner wrote. 


```python
# Partner's code
from typing import List

def missing_num(nums: List[int]) -> List[int]:
    n = max(nums)  
    num_set = set(nums)
    missing = []

    for i in range(n + 1):
        if i not in num_set:
            missing.append(i)

    return missing if missing else [-1]

# Example 1
nums1 = [0,8,4,12,2,3]
print("Example 1:")
print("Input:", nums1)
print("Output:", missing_num(nums1))

# Example 2
nums2 = [9,6,7,1]
print("\nExample 2:")
print("Input:", nums2)
print("Output:", missing_num(nums2))
```
